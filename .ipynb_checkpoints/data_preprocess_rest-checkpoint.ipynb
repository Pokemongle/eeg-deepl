## 2.2. Load the data

```python
import gc
gc.collect()
torch.cuda.empty_cache()

# GPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

dataset = MyData(root_dir,exper_dir,condi_dir)
for i in tqdm(range(len(dataset))): #121
    # filepath.mat
    # path.mat
    filename = os.path.join(dataset.path,dataset.file_path[i])
    data = scipy.io.loadmat(filename)#
    data = data['datas']
    
    # GPU
    data = torch.from_numpy(data).to(device)
    
    # GPU
    data_map_all = None # 
    for para in tqdm(range(data.shape[2])): # 
        data_map = None # 
        for point in tqdm(range(data.shape[1])):
            map = torch.zeros((10, 11)).to(device)
            for channel in tqdm(range(data.shape[0])):
                map[matrix_index[0][channel]][matrix_index[1][channel]] = data[channel][point][para]
            if data_map is None:
                data_map = map.unsqueeze(0)
            else:
                data_map = torch.cat((data_map, map.unsqueeze(0)), dim=0)
        if data_map_all is None:
            data_map_all = data_map.unsqueeze(0)
        else:
            data_map_all = torch.cat((data_map_all, data_map.unsqueeze(0)), dim=0)
    gc.collect()
    torch.cuda.empty_cache()
```
